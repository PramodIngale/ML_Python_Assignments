```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
```

```python
#import dataset
salary_train = pd.read_csv("SalaryData_Train.csv")
salary_test = pd.read_csv("SalaryData_Test.csv")

salary_train.head()
```

### EDA

```python
#NA values
salary_train.info()
```

```python
#Duplicate rows
salary_train[salary_train.duplicated()]
```

```python
#Delete duplicate rows
salary_train.drop_duplicates(inplace=True, ignore_index=True)
```

```python
#descriptive statistics
salary_train.describe()
```

```python
#Data preparation for model
x_train = pd.get_dummies(salary_train.iloc[:,0:-1])
y_train = LabelEncoder().fit_transform(salary_train.iloc[:,-1])

x_test = pd.get_dummies(salary_test.iloc[:,0:-1])
y_test = LabelEncoder().fit_transform(salary_test.iloc[:,-1])
```

```python
x_test
```

### Model

```python
#Multinomial naive bayes model
model_MNB = MultinomialNB()
model_MNB.fit(x_train.iloc[:,5:],y_train) #Supplied only catogorical value
round(model_MNB.score(x_train.iloc[:,5:],y_train),3) #Supplied only catogorical value
```

```python
#model score on test data
round(model_MNB.score(x_test.iloc[:,5:],y_test),3) #Supplied only catogorical value
```

```python
#Guassian naive bayes model
model_GNB = GaussianNB()
model_GNB.fit(x_train.iloc[:,5:],y_train) #Supplied only catogorical value
round(model_GNB.score(x_train.iloc[:,5:],y_train),3) #Supplied only catogorical value
```

```python
#model score on test data
round(model_GNB.score(x_test.iloc[:,5:],y_test),3) #Supplied only catogorical value
```
