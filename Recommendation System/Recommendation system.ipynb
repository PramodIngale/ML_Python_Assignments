```python
#import libraries
import pandas as pd
import numpy as np

from sklearn.metrics import pairwise_distances
```

```python
#import dataset
book_df = pd.read_csv("book.csv",encoding='ISO-8859-1',index_col=0)
book_df.head()
```

## EDA

```python
#Duplicate redords
book_df[book_df.iloc[:,0:2].duplicated()]
```

```python
#Delete the duplicate records
book_df.drop(book_df[book_df.iloc[:,0:2].duplicated()].index, inplace=True)
book_df.reset_index(drop=True,inplace=True)
```

```python
#total number of records
book_df.shape
```

```python
#Total number of unique users
len(book_df["User.ID"].unique())
```

```python
#Total number of unique books
len(book_df["Book.Title"].unique())
```

```python
#Create pivot table using userID and booktitle
user_book_df = book_df.pivot(index="User.ID",
                                 columns='Book.Title',
                                 values='Book.Rating')
```

```python
#Fill na values with 0
user_book_df.fillna(0,inplace=True)
```

```python
user_book_df
```

## Similarity Calculation

```python
#cosine similarity
user_sim = 1- pairwise_distances(user_book_df.values, metric="cosine")
user_sim
```

```python
#Dataframe with cosine scores
np.fill_diagonal(user_sim,0)
user_sim_df = pd.DataFrame(user_sim)
user_sim_df.index = user_book_df.index
user_sim_df.columns = user_book_df.index
user_sim_df
```

```python
#Now similar users can found from the above datframe, example
user_sim_df.idxmax(axis=1)[0:15]
```
